```python
questions = ["1. How many people have heart disease and how many people doesn't have disease? ",
             "2. People of which sex has most heart disease? ",
             "3. People of which sex has which type of chest pain most? ",
             "4. People with which chest pain or most pron to have heart disease?"]
questions
```

```python
#let's find the first answer of the first question

#how many people have heart disease and how many people doesn't have heart disease?

#getting the values
df.target.value_counts()
```

```python
#plotting bar chart
df.target.value_counts().plot(kind = 'bar', color = ["orchid","salmon"])
plt.title("heart disease vslues")
plt.xlabel("1 = heart disease,0 = no heart disease")
plt.ylabel("amount");
```

```python
#plotting a pie chart
df.target.value_counts().plot(kind = "pie",figsize = (8,6))
plt.legend({"disease","no disease"});
```

```python
# '0' represent female
# '1' represent male
#sex column part
# '0' represent disease
# '1' represnt no disease
#target column part
#now let's check how many 'male' amd 'female' are in the dataset
df.sex.value_counts()
```

```python
#plotting a pie chart
df.sex.value_counts().plot(kind = "pie",figsize = (8,6))
plt.title("Male Female ratio ")
plt.legend({"Male","Female)"});
```

```python
#let's find the answer of our 2nd question
#2.people of which swx has most heart disease?
pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x = "target",data = df,hue = "sex")
plt.title("heart disease frequency for swx")
plt.xlabel("0 = no heart disease, 1 = heart disease");
```

```python
#Number of males is more than double in our dataset than female.

#More than '45% Male' has heart diseasw and '75% female' has heart disease.
```

```python
#Let's move to question 3

#3. 'people of which sex has which type of chest pain most?'

#counting values for different chest pain
df.cp.value_counts()
```

```python
#plotting a bar chart
df.cp.value_counts().plot(kind = 'bar', color = ['salmon','lightskyblue','springgreen','khaki'])
plt.title('Chest pain type vs count');
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind = "bar", color = ['salmon','lightskyblue','springgreen','khaki'])
plt.title('Type of chest pain vs sex')
plt.xlabel('0 = Female, 1 = Male');
```

```python
# Most of 'male' has 'type 0' chest pain and least 'Male'has 'type 4' pain.

# In case of 'Female' 'type 0' and 'type 2' Percentage is almost same.
```

```python
#Now question 4

#4.people with which chest pain or most pron to have heart disease?

pd.crosstab(df.cp, df.target)
```

```python
sns.countplot(x = 'cp',data = df,hue = 'target');
```

```python
#Most of the people has 'type 0' chest pain and has less chance of heart disease.

#And we see the opposite for other types.

#Now lets take a look at our age column.

#Create a distribution plot with normal distribution curve
sns.displot(x = 'age', data = df, bins = 30, kde = True);
```
